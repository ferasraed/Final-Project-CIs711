# Dim 10_01

```python
import pandas as pd
from scipy.stats import friedmanchisquare

file_path = 'Dim 10_1.csv'
data = pd.read_csv(file_path)

data_papers = data.iloc[:, 1:]

statistic, p_value = friedmanchisquare(*[data_papers[paper] for paper in data_papers.columns])

print(f"Friedman test statistic: {statistic}")
print(f"p-value: {p_value}")

ranks = data_papers.rank(axis=1)

average_ranks = ranks.mean()

average_ranks_sorted = average_ranks.sort_values(ascending=False)
print("\nAverage ranks for each paper:")
print(average_ranks_sorted)
```

* Ranking the Data: Within each block (subject), the data is ranked. Ties receive the average of the ranks they would have received.
* Sum of Ranks: The ranks for each treatment are summed across all blocks.



n: entites (functions). This corresponds to the number of rows in your dataset.

k: The groups, This corresponds to the number of columns in your dataset.

Rj:The sum of ranks. This is calculated by ranking the data within each block and then summing these ranks for each treatment.


## Friedman average ranking results

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('Dim 10_1.csv')

data_papers = data.iloc[:, 1:]  # Assuming data is already loaded and prepared
ranks = data_papers.rank(axis=1)
average_ranks = ranks.mean()
average_ranks_sorted = average_ranks.sort_values(ascending=False)

colors = ['#4D79DA', '#EF8133', '#B8B8B8', '#FFCF00', '#77ADD4', '#52B242', '#1D348A', '#9B3F00', '#716C6B', '#A37503', '#1A3B6F']

plt.figure(figsize=(10, 6))  # Set the figure size

average_ranks_sorted.plot(kind='bar', color=colors[:len(average_ranks_sorted)])
plt.title('Average Ranks of Papers - Friedman Test Results')
plt.xlabel('Papers')
plt.ylabel('Average Rank')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# **Wilcoxon Signed-Rank Test**

## Results according on a control method p10

# **Dim 10**

# Dim 10_01

```python
import pandas as pd
from scipy.stats import wilcoxon

means = pd.read_csv('Dim 10_1.csv', index_col=0)
stds = pd.read_csv('Dim 10_std_1.csv', index_col=0)

functions = means.index.tolist()
papers = means.columns.tolist()

reference_paper = 'p10'

if reference_paper not in papers:
    papers.append(reference_paper)

formatted_results = pd.DataFrame(index=functions, columns=pd.MultiIndex.from_product([papers, ['F', 'STD', '+/-/=']]))

for function in functions:
    formatted_results.loc[function, (reference_paper, 'F')] = means.loc[function, reference_paper]
    formatted_results.loc[function, (reference_paper, 'STD')] = stds.loc[function, reference_paper]
    formatted_results.loc[function, (reference_paper, '+/-/=')] = '='  # Indicate no comparison

for paper in papers:
    if paper == reference_paper:
        continue  # Skip comparison logic for p10
    for function in functions:
        data_ref = means.loc[function, reference_paper]
        data_comp = means.loc[function, paper]
        std_ref = stds.loc[function, reference_paper]
        std_comp = stds.loc[function, paper]

        try:
            stat, p_value = wilcoxon(means.loc[:, reference_paper].dropna(), means.loc[:, paper].dropna(), alternative='two-sided')
            if p_value <= 0.05:
                symbol = '-' if data_ref > data_comp else '+'
            else:
                symbol = '='
        except Exception as e:
            symbol = 'Err'
            print(f"Error: {e}")

        formatted_results.loc[function, (paper, 'F')] = data_comp
        formatted_results.loc[function, (paper, 'STD')] = std_comp
        formatted_results.loc[function, (paper, '+/-/=')] = symbol

formatted_results.to_csv('formatted_wilcoxon_results_control_p10_1.csv')
print(formatted_results.head())
```

## Counting superior performances

```python
import pandas as pd
from scipy.stats import wilcoxon

means = pd.read_csv('Dim 10_14.csv', index_col=0)
stds = pd.read_csv('Dim 10_std_14.csv', index_col=0)

functions = means.index.tolist()
papers = means.columns.tolist()

reference_paper = 'p10'

if reference_paper not in papers:
    papers.append(reference_paper)

formatted_results = pd.DataFrame(index=functions, columns=pd.MultiIndex.from_product([papers, ['F', 'STD', '+/-/=']]))

for function in functions:
    formatted_results.loc[function, (reference_paper, 'F')] = means.loc[function, reference_paper]
    formatted_results.loc[function, (reference_paper, 'STD')] = stds.loc[function, reference_paper]
    formatted_results.loc[function, (reference_paper, '+/-/=')] = '='  # Indicate no comparison

for paper in papers:
    if paper == reference_paper:
        continue  # Skip comparison logic for p10
    for function in functions:
        data_ref = means.loc[function, reference_paper]
        data_comp = means.loc[function, paper]
        std_ref = stds.loc[function, reference_paper]
        std_comp = stds.loc[function, paper]

        try:
            stat, p_value = wilcoxon(means.loc[:, reference_paper].dropna(), means.loc[:, paper].dropna(), alternative='two-sided')
            if p_value <= 0.05:
                symbol = '-' if data_ref > data_comp else '+'
            else:
                symbol = '='
        except Exception as e:
            symbol = 'Err'
            print(f"Error: {e}")

        formatted_results.loc[function, (paper, 'F')] = data_comp
        formatted_results.loc[function, (paper, 'STD')] = std_comp
        formatted_results.loc[function, (paper, '+/-/=')] = symbol

formatted_results.to_csv('formatted_wilcoxon_results_control_p10_14.csv')
print(formatted_results.head())
```

## Counting superior performances

```python
data_wilcoxon = pd.read_csv('formatted_wilcoxon_results_control_p10_14.csv')
print(data_wilcoxon.head())

results = {}
comparison_columns = [col for col in data_wilcoxon.columns if col.endswith('.2')]

for col in comparison_columns:
    paper = col.split('.')[0]
    results[paper] = 0

for index, row in data_wilcoxon.iterrows():
    for col in comparison_columns:
        paper = col.split('.')[0]
        symbol = row[col]
        if symbol == '-':
            results[paper] += 1

print("Number of times each paper was better than the control (p10):")
for paper, count in results.items():
    print(f"{paper}: {count}")
```
